```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Get only unique players
unique_players = purchase_data.drop_duplicates("SN")

# Player Count
count = unique_players["SN"].count()
player_count = [{"Total Players" : count}]
player_count_df = pd.DataFrame(player_count)
player_count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate Number of Unique Items
unique_items = purchase_data.drop_duplicates("Item ID")
number_of_unique_items = unique_items["Item ID"].count()

# Calculate Average Price
average_price = unique_items["Price"].mean()

# Get Number of Purchases
number_of_purchases = purchase_data["Item ID"].count()

# Total Revenue
total_revenue = purchase_data["Price"].sum()

# Creating Purchase Analysis (Total)
purchase_analysis = [{"Number of Unique Items" : number_of_unique_items, 
                      "Average Price" : average_price, 
                      "Number of Purchases": number_of_purchases,
                      "Total Revenue" : total_revenue
                     }]
purchase_analysis_df = pd.DataFrame(purchase_analysis)

# # Formatting Purchase Analysis
purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map("${:.2f}".format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map("${:,.2f}".format)
purchase_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Get only unique players
unique_players = purchase_data.drop_duplicates("SN")

# Total Gender Count
gender_demographics= unique_players["Gender"].value_counts()
gender_demographics_df = pd.DataFrame(gender_demographics)
gender_demographics_df = gender_demographics_df.rename(columns={"Gender":"Total Count"})

# Count of Total Players
total = gender_demographics_df["Total Count"].sum()

#Percentage of Players
percentage_players = gender_demographics_df["Total Count"]/total * 100 
gender_demographics_df["Percentage of Players"] = percentage_players
gender_demographics_df["Percentage of Players"] =gender_demographics_df["Percentage of Players"].map("{0:.2f}%".format)
gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchase Count per Gender
gender_purchase = purchase_data["Gender"].value_counts()
gender_purchase_df = pd.DataFrame(gender_purchase)
gender_purchase_df = gender_purchase_df.rename(columns={"Gender":"Purchase Count"})

# Male only
male_only = purchase_data.loc[purchase_data["Gender"] == "Male",:]
male_count = male_only["SN"].count()
male_avg_price = male_only["Price"].mean()
male_value = male_only["Price"].sum()
male_group = male_only.groupby("SN")
male_avg_total = male_group["Price"].sum()
male_avg_total = male_avg_total.mean()    

# Female only
female_only = purchase_data.loc[purchase_data["Gender"] == "Female",:]
female_count = female_only["SN"].count()
female_avg_price = female_only["Price"].mean()
female_value = female_only["Price"].sum()
female_group = female_only.groupby("SN")
female_avg_total = female_group["Price"].sum()
female_avg_total = female_avg_total.mean()    

# Other
other_only = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed",:]
other_count = other_only["SN"].count()
other_avg_price = other_only["Price"].mean()
other_value = other_only["Price"].sum()
other_group = other_only.groupby("SN")
other_avg_total = other_group["Price"].sum()
other_avg_total = other_avg_total.mean()    


# Analysis by Gender
gender_analysis_df = pd.DataFrame(
    {"Gender": ["Male","Female","Other / Non-Disclosed"],
    "Purchase Count":[male_count,female_count,other_count],
    "Average Purchase Price":[male_avg_price,female_avg_price,other_avg_price],
    "Total Purchase Value": [male_value,female_value,other_value],
    "Avg Total Purchase Per Person": [male_avg_total,female_avg_total,other_avg_total]})
#Formatting
gender_analysis_df["Average Purchase Price"]= gender_analysis_df["Average Purchase Price"].map("${:,.2f}".format)
gender_analysis_df["Total Purchase Value"] = gender_analysis_df["Total Purchase Value"].map("${:,.2f}".format)
gender_analysis_df["Avg Total Purchase Per Person"] = gender_analysis_df["Avg Total Purchase Per Person"].map("${:,.2f}".format)
gender_analysis_df= gender_analysis_df.set_index("Gender")

gender_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create bins
bins = [0, 9, 14, 19, 24, 29, 34, 40, 100]
age_range =["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_bin = pd.cut(unique_players["Age"], bins, labels=age_range, include_lowest=True)
age_bin = pd.DataFrame(age_bin)
age_bin = age_bin.groupby("Age")
age_bin= age_bin.max()

# Count Unique players
count_unique_players = unique_players["SN"].count()

#<10
age_group_one = unique_players.loc[unique_players["Age"] < 10, ["Age"]]
count_group_one = age_group_one.count()
group_one_pc = count_group_one/count_unique_players * 100

#10-14
age_group_two = unique_players.loc[(unique_players["Age"] >= 10) & (unique_players["Age"] <=14), ["Age"]]
count_group_two = age_group_two.count()
group_two_pc = count_group_two/count_unique_players * 100

# #15-19
age_group_three = unique_players.loc[(unique_players["Age"] >= 15) & (unique_players["Age"] <=19), ["Age"]]
count_group_three = age_group_three.count()
group_three_pc = count_group_three/count_unique_players * 100

#20-24
age_group_four = unique_players.loc[(unique_players["Age"] >= 20) & (unique_players["Age"] <=24), ["Age"]]
count_group_four = age_group_four.count()
group_four_pc = count_group_four/count_unique_players *100

#25-29
age_group_five = unique_players.loc[(unique_players["Age"] >= 25) & (unique_players["Age"] <=29), ["Age"]]
count_group_five = age_group_five.count()
group_five_pc = count_group_five/count_unique_players * 100

#30-34
age_group_six = unique_players.loc[(unique_players["Age"] >= 30) & (unique_players["Age"] <=34), ["Age"]]
count_group_six = age_group_six.count()
group_six_pc = count_group_six/count_unique_players * 100

#35-39
age_group_seven = unique_players.loc[(unique_players["Age"] >= 35) & (unique_players["Age"] <=39), ["Age"]]
count_group_seven = age_group_seven.count()
group_seven_pc = count_group_seven/count_unique_players * 100

#40+
age_group_eight = unique_players.loc[(unique_players["Age"] >= 40), ["Age"]]
count_group_eight = age_group_eight.count()
group_eight_pc = count_group_eight/count_unique_players * 100

#Create Dataframes
range_age = [count_group_one, count_group_two, count_group_three, count_group_four,
            count_group_five, count_group_six, count_group_seven,count_group_eight]
range_age = pd.DataFrame(range_age)
pc_age = [group_one_pc, group_two_pc, group_three_pc, group_four_pc,
          group_five_pc, group_six_pc, group_seven_pc, group_eight_pc]
pc_age = pd.DataFrame(pc_age)

age_demographics = {"Age":["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]}
age_demographics = pd.DataFrame(age_demographics)
age_demographics["Total Count"] = range_age
age_demographics["Percentage of Players"] = pc_age
age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{0:.2f}%".format)

#Merge to bin
age_demographics = pd.merge(age_bin, age_demographics, on= "Age")
age_demographics = age_demographics.set_index("Age")
age_demographics.index.name = None
age_demographics
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Purchase Count
Spender_count = purchase_data["SN"].value_counts()

#Group by SN
Sorted_purchase_data = purchase_data.groupby("SN")

#Average Purchase Price
Spender_Avg = Sorted_purchase_data["Price"].mean()

#Total Purchase Price
Spender_Total = Sorted_purchase_data["Price"].sum()

#Top Spenders
Top_Spenders = pd.merge(Spender_Avg, Spender_Total, on= "SN")
Top_Spenders["Purchase Count"] = Spender_count
Top_Spenders = Top_Spenders.rename(columns={  "Price_x": "Avg Purchase Price", "Price_y" :"Total Purchase Value"})
Top_Spenders = Top_Spenders[["Purchase Count", "Avg Purchase Price", "Total Purchase Value"]]

#Format
Top_Spenders["Avg Purchase Price"] = Top_Spenders["Avg Purchase Price"].map("${:,.2f}".format)
Top_Spenders["Total Purchase Value"] = Top_Spenders["Total Purchase Value"].map("${:,.2f}".format)

#Sort by Total Purchase value
Top_Spenders = Top_Spenders.sort_values(["Purchase Count","Total Purchase Value"], ascending=False)


#Display top 5
Top_Spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group By
Grouped_Item_data = purchase_data.groupby(["Item ID","Item Name"])

#Item Count
Item_Count = Grouped_Item_data.count()
Item_Count = Item_Count.rename(columns={"Purchase ID": "Purchase Count"})
Item_Count = Item_Count["Purchase Count"]

#Get Price
Item_Price = Grouped_Item_data.max()
Item_Price = Item_Price[["Price"]]
Item_Price = Item_Price.rename(columns={"Price": "Item Price"})
Item_Price["Item Price"] = Item_Price["Item Price"].map("${:,.2f}".format)

#Total Purchase Value
Total_Purchase_Value = pd.DataFrame(Grouped_Item_data["Price"].sum())
Total_Purchase_Value = Total_Purchase_Value.rename(columns={"Price" : "Total Purchase Value"})
Total_Purchase_Value["Total Purchase Value"] = Total_Purchase_Value["Total Purchase Value"].map("${:,.2f}".format)

#Merge
Popular_Items = pd.merge(Item_Count, Item_Price, left_index= True, right_index= True)
Popular_Items = pd.merge(Popular_Items, Total_Purchase_Value, left_index= True, right_index= True)

Most_Popular_Items = Popular_Items.sort_values(["Purchase Count"], ascending= False)
Most_Popular_Items.head()
```
